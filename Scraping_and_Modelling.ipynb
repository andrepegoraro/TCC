## Automatizando o Código

```python
import requests
import pandas as pd

# Definição dos cabeçalhos sem cabeçalhos de cache
headers = {
    'accept': '*/*',
    'accept-language': 'en-US,en;q=0.9',
    'cache-control': 'no-cache',
    'pragma': 'no-cache',
    'referer': 'https://www.sofascore.com/',
    'sec-ch-ua': '" Not;A Brand";v="99", "Google Chrome";v="91", "Chromium";v="91"',
    'sec-ch-ua-mobile': '?0',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-origin',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'x-sofa-api-version': '1.0',
}

# Lista de URLs
urls = [
    "https://www.sofascore.com/api/v1/event/10113678/shotmap",
    "https://www.sofascore.com/api/v1/event/10114121/shotmap",
    "https://www.sofascore.com/api/v1/event/10113769/shotmap",
    "https://www.sofascore.com/api/v1/event/10113772/shotmap",
    "https://www.sofascore.com/api/v1/event/10113936/shotmap",
    "https://www.sofascore.com/api/v1/event/10114016/shotmap",
    "https://www.sofascore.com/api/v1/event/10113771/shotmap",
    "https://www.sofascore.com/api/v1/event/10114108/shotmap",
    "https://www.sofascore.com/api/v1/event/10113978/shotmap",
    "https://www.sofascore.com/api/v1/event/10113910/shotmap"
]

# Lista para armazenar todos os dados de chutes
all_shots = []

# Loop para iterar sobre cada URL
for url in urls:
    response = requests.get(url, headers=headers)

    # Verificar se a requisição foi bem-sucedida
    if response.status_code == 200:
        data = response.json()

        # Extrair os dados dos chutes e armazenar
        for shot in data['shotmap']:
            shot_info = {
                'event_id': url.split('/')[-2],  # Adiciona o ID do evento para referência
                'player_name': shot['player']['name'],
                'player_position': shot['player'].get('position', ''),
                'player_jersey_number': shot['player'].get('jerseyNumber', ''),
                'player_id': shot['player']['id'],
                'is_home': shot['isHome'],
                'shot_type': shot['shotType'],
                'situation': shot['situation'],
                'player_x': shot['playerCoordinates']['x'],
                'player_y': shot['playerCoordinates']['y'],
                'player_z': shot['playerCoordinates'].get('z', 0),
                'body_part': shot['bodyPart'],
                'goal_mouth_location': shot['goalMouthLocation'],
                'goal_mouth_x': shot['goalMouthCoordinates']['x'],
                'goal_mouth_y': shot['goalMouthCoordinates']['y'],
                'goal_mouth_z': shot['goalMouthCoordinates']['z'],
                'xg': shot.get('xg', None),
                'shot_id': shot['id'],
                'time': shot['time'],
                'added_time': shot.get('addedTime', 0),
                'time_seconds': shot['timeSeconds'],
                'incident_type': shot['incidentType']
            }
            all_shots.append(shot_info)
    else:
        print(f"Erro na requisição para {url}: {response.status_code}")

# Convertendo todos os dados de chutes para um DataFrame
dataframe = pd.DataFrame(all_shots)

# Exibindo os primeiros 10 registros
print(dataframe.head(10))
dataframe.info()

# Salvar o DataFrame em um arquivo CSV, se necessário
# df.to_csv('all_shotmaps.csv', index=False)
```
